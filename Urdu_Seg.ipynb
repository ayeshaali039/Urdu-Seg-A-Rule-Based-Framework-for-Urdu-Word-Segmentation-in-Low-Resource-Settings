```python
import pandas as pd
import re

def urdu_tokenizer(text):
    """
    Tokenize Urdu text using regex.
    Keeps Urdu letters, numbers, and basic punctuation.
    """
    if not isinstance(text, str):
        return []

    # Normalize spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Regex pattern for Urdu words, numbers, and punctuation
    pattern = r'[\u0600-\u06FF]+|\d+|[.,!?؛،]'
    
    # Find all tokens
    tokens = re.findall(pattern, text)
    return tokens

# Load CSV file
file_path = "token.csv"
df = pd.read_csv(file_path)

if 'comment' not in df.columns:
    raise ValueError("CSV file must contain a 'comment' column")

# Apply tokenizer
df['tokens'] = df['comment'].apply(urdu_tokenizer)

# Convert list of tokens to string representation like Python list
df['tokens'] = df['tokens'].apply(lambda x: str(x))

# Save output in CSV with UTF-8 encoding for Urdu readability
output_file = "regex.csv"
df.to_csv(output_file, index=False, encoding='utf-8-sig')

print(f"Tokenization completed. Output saved to {output_file}")
```

```python
import pandas as pd
import ast
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# CSV files for different tokenizers
FILES = {
    "Whitespace": "whitespace.csv",
    "Regex": "regex.csv",
    "spaCy": "spacy.csv"
}

TOKENS_COLUMN = "tokens"

results = []

for name, path in FILES.items():
    print(f"\nProcessing tokenizer: {name}")
    df = pd.read_csv(path)
    
    # Convert string representation of tokens to list
    df[TOKENS_COLUMN] = df[TOKENS_COLUMN].apply(ast.literal_eval)
    
    # Reconstruct text from tokens
    df['text'] = df[TOKENS_COLUMN].apply(lambda x: " ".join(x))
    
    # TF-IDF vectorization
    vectorizer = TfidfVectorizer(
        ngram_range=(1, 2),
        min_df=2,
        max_df=0.95
    )
    X_vec = vectorizer.fit_transform(df['text'])
    
    # KMeans clustering
    n_clusters = 5  # choose number of clusters; adjust as needed
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_vec)
    
    # Silhouette score
    if len(set(cluster_labels)) > 1:
        score = silhouette_score(X_vec, cluster_labels)
    else:
        score = np.nan  # cannot compute if only 1 cluster
    
    # Basic intrinsic metrics
    token_counts = df[TOKENS_COLUMN].apply(len)
    vocab = set(token for tokens in df[TOKENS_COLUMN] for token in tokens)
    
    results.append({
        "Tokenizer": name,
        "NumComments": len(df),
        "AvgTokensPerComment": round(np.mean(token_counts), 2),
        "MinTokensPerComment": int(np.min(token_counts)),
        "MaxTokensPerComment": int(np.max(token_counts)),
        "VocabularySize": len(vocab),
        "SilhouetteScore": round(score, 4)
    })

# Create DataFrame with results
results_df = pd.DataFrame(results)
print("\nTokenizer Comparison Results:\n")
print(results_df)

# Save results
results_df.to_csv("tokenizer_comparison_results.csv", index=False, encoding="utf-8-sig")
```

```python

import spacy
import stanza
from nltk.tokenize import wordpunct_tokenize
#from indicnlp.tokenize import indic_tokenize

# ------------------------------
# Load data
# ------------------------------
def load_lines(path):
    with open(path, "r", encoding="utf-8") as f:
        return [line.rstrip("\n") for line in f]

import pandas as pd
raw = load_lines("data_tokenization_5000.csv")
gold_lines = load_lines("gold_standard_cleaned.csv")
spacy_custom_lines = load_lines("spacy_tokenized_output.csv")

gold_tokens = [l.split() for l in gold_lines]
spacy_custom_tokens = [l.split() for l in spacy_custom_lines]

# ------------------------------
# Tokenizers
# ------------------------------
nlp_spacy = spacy.blank("ur")

def spacy_default(text):
    return [t.text for t in nlp_spacy(text)]

def nltk_tok(text):
    return wordpunct_tokenize(text)

stanza.download("ur", verbose=False)
nlp_stanza = stanza.Pipeline("ur", processors="tokenize", tokenize_no_ssplit=True, verbose=False)

def stanza_tok(text):
    doc = nlp_stanza(text)
    return [w.text for s in doc.sentences for w in s.words]

#def indic_tok(text):
   # return indic_tokenize.trivial_tokenize(text)

def whitespace_tok(text):
    return text.split()

tokenizers = {
    "Urdu_Seg": spacy_custom_tokens,
    "spaCy_default": [spacy_default(s) for s in raw],
    "NLTK": [nltk_tok(s) for s in raw],
    "Stanza": [stanza_tok(s) for s in raw],
   # "IndicNLP": [indic_tok(s) for s in raw],
    "Whitespace": [whitespace_tok(s) for s in raw],
}

# ------------------------------
# Convert tokens to boundaries
# ------------------------------
def to_offsets(text, tokens):
    offsets = []
    pos = 0
    for tok in tokens:
        idx = text.find(tok, pos)
        if idx == -1:
            continue
        offsets.append((idx, idx + len(tok)))
        pos = idx + len(tok)
    return offsets

# ------------------------------
# Evaluate
# ------------------------------
def evaluate(gold, pred):
    gold = set(gold)
    pred = set(pred)

    TP = len(gold & pred)
    FP = len(pred - gold)
    FN = len(gold - pred)

    return TP, FP, FN

# ------------------------------
# Run evaluation
# ------------------------------
results = {}

for name, preds in tokenizers.items():
    TP = FP = FN = 0

    for i in range(len(raw)):
        gold_offsets = to_offsets(raw[i], gold_tokens[i])
        pred_offsets = to_offsets(raw[i], preds[i])

        tp, fp, fn = evaluate(gold_offsets, pred_offsets)
        TP += tp
        FP += fp
        FN += fn

    precision = TP / (TP + FP + 1e-9)
    recall = TP / (TP + FN + 1e-9)
    f1 = 2 * precision * recall / (precision + recall + 1e-9)
    accuracy = TP / (TP + FP + FN + 1e-9)

    results[name] = (precision, recall, f1, accuracy, TP, FP, FN)

# ------------------------------
# Print results
# ------------------------------
print("\nBOUNDARY-BASED TOKENIZATION RESULTS\n")
print("{:<15} {:<9} {:<9} {:<9} {:<9}".format("Tokenizer", "Prec", "Recall", "F1", "Acc"))
print("-" * 55)

for name, (p, r, f, a, _, _, _) in results.items():
    print("{:<15} {:.4f}   {:.4f}   {:.4f}   {:.4f}".format(name, p, r, f, a))

print("\nCONFUSION VALUES (TP, FP, FN)\n")
for name, (_, _, _, _, TP, FP, FN) in results.items():
    print(name, "→", {"TP": TP, "FP": FP, "FN": FN, "TN": 0})
```

```python
import spacy
import stanza
from nltk.tokenize import wordpunct_tokenize

# ------------------------------
# Load data
# ------------------------------
def load_lines(path):
    with open(path, "r", encoding="utf-8") as f:
        return [line.rstrip("\n") for line in f]

raw = load_lines("data_tokenization_500.csv")
gold_lines = load_lines("gold_standard_cleaned.csv")
spacy_custom_lines = load_lines("spacy_tokenized_output.csv")

gold_tokens = [l.split() for l in gold_lines]
spacy_custom_tokens = [l.split() for l in spacy_custom_lines]

# ------------------------------
# Tokenizers
# ------------------------------
nlp_spacy = spacy.blank("ur")

def spacy_default(text):
    return [t.text for t in nlp_spacy(text)]

def nltk_tok(text):
    return wordpunct_tokenize(text)

stanza.download("ur", verbose=False)
nlp_stanza = stanza.Pipeline("ur", processors="tokenize",
                             tokenize_no_ssplit=True,
                             verbose=False)

def stanza_tok(text):
    doc = nlp_stanza(text)
    return [w.text for s in doc.sentences for w in s.words]

def whitespace_tok(text):
    return text.split()

tokenizers = {
    "Rule_based": spacy_custom_tokens,
    "spaCy_default": [spacy_default(s) for s in raw],
    "NLTK": [nltk_tok(s) for s in raw],
    "Stanza": [stanza_tok(s) for s in raw],
    "Whitespace": [whitespace_tok(s) for s in raw],
}

# ------------------------------
# Similarity functions
# ------------------------------
def jaccard_similarity(gold, pred):
    gold_set = set(gold)
    pred_set = set(pred)

    union = gold_set | pred_set
    if len(union) == 0:
        return 1.0

    return len(gold_set & pred_set) / len(union)

# ------------------------------
# Run similarity evaluation
# ------------------------------
results = {}

for name, preds in tokenizers.items():
    sim_total = 0.0

    for i in range(len(raw)):
        sim = jaccard_similarity(gold_tokens[i], preds[i])
        sim_total += sim

    avg_similarity = sim_total / len(raw)
    results[name] = avg_similarity

# ------------------------------
# Print results
# ------------------------------
print("\nTOKENIZATION SIMILARITY RESULTS (Jaccard)\n")
print("{:<15} {:<10}".format("Tokenizer", "Jaccard"))
print("-" * 30)

for name, sim in results.items():
    print("{:<15} {:.4f}".format(name, sim))
```

```python
import pandas as pd
import spacy
import re
import unicodedata
from spacy.tokenizer import Tokenizer
from spacy.util import compile_prefix_regex, compile_suffix_regex, compile_infix_regex
from spacy.symbols import ORTH

# ===============================
# Load Dataset
# ===============================
data = pd.read_csv("data_tokenization_500.csv", encoding="utf-8")

print("📘 Dataset preview:")
print(data.head())

text_column = "comment"   # 🔹 Change if your column name is different


# ===============================
# Rule 1: Urdu-only Script + Unicode Normalization
# ===============================
def normalize_urdu(text):
    text = unicodedata.normalize("NFC", str(text))
    text = re.sub(r"[^\u0600-\u06FF\s]", " ", text)  # Keep only Urdu script
    text = re.sub(r"\s+", " ", text).strip()
    return text


# ===============================
# Initialize spaCy Urdu Tokenizer
# ===============================
nlp = spacy.blank("ur")


# ===============================
# Rule 2: Prefix Recognition
# ===============================
urdu_prefixes = ["بے", "نا", "بد", "لا"]

prefix_patterns = list(nlp.Defaults.prefixes) + [
    r"^" + p for p in urdu_prefixes
]

prefix_regex = compile_prefix_regex(prefix_patterns)


# ===============================
# Rule 3: Suffix / Clitic Separation
# ===============================
urdu_suffixes = ["کو", "سے", "نے", "تک", "ہی", "بھی"]

suffix_patterns = list(nlp.Defaults.suffixes) + [
    r"" + s + r"$" for s in urdu_suffixes
]

suffix_regex = compile_suffix_regex(suffix_patterns)


# ===============================
# Rule 5: Infix / Punctuation Handling
# ===============================
urdu_punct = [r"[،۔!?؛]"]

infix_patterns = list(nlp.Defaults.infixes) + urdu_punct

infix_regex = compile_infix_regex(infix_patterns)


# ===============================
# Create Custom Tokenizer
# ===============================
nlp.tokenizer = Tokenizer(
    nlp.vocab,
    prefix_search=prefix_regex.search,
    suffix_search=suffix_regex.search,
    infix_finditer=infix_regex.finditer,
)


# ===============================
# Rule 4: Compound & Named Entity Preservation
# ===============================
compound_phrases = [
    "وزیراعظم پاکستان",
    "اسلام آباد",
    "سپریم کورٹ",
    "اہل خانہ"
]

for phrase in compound_phrases:
    nlp.tokenizer.add_special_case(
        phrase,
        [{ORTH: phrase}]
    )


# ===============================
# Tokenization Function
# ===============================
def tokenize_text(text):
    text = normalize_urdu(text)  # Apply Rule 1
    doc = nlp(text)              # Apply Rules 2–5
    return [token.text for token in doc]


# ===============================
# Apply Tokenization
# ===============================
data["tokens"] = data[text_column].apply(tokenize_text)

print("✅ Tokenization complete!")
print(data.head())

# ===============================
# Save Output
# ===============================
data.to_csv("spacy_tokenized_output.csv", index=False, encoding="utf-8-sig")

print("💾 Saved as spacy_tokenized_output.csv")
```

```python
import pandas as pd
import ast

# ===============================
# Load Files
# ===============================
pred = pd.read_csv("spacy_tokenized_output.csv", encoding="utf-8")
gold = pd.read_csv("gold_standard.csv", encoding="utf-8")

# Check columns
print("Gold columns:", gold.columns)
print("Pred columns:", pred.columns)

# ===============================
# Convert string to list safely
# ===============================
def to_list(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    return x

pred["tokens"] = pred["tokens"].apply(to_list)
gold["tokens"] = gold["manually_tokenized"].apply(to_list)

# ===============================
# Jaccard Similarity
# ===============================
def jaccard_similarity(list1, list2):
    set1 = set(list1)
    set2 = set(list2)

    intersection = len(set1 & set2)
    union = len(set1 | set2)

    return intersection / union if union != 0 else 0

# ===============================
# Compute Safely
# ===============================
min_length = min(len(gold), len(pred))

jaccard_scores = []

for i in range(min_length):
    score = jaccard_similarity(
        gold.iloc[i]["tokens"],
        pred.iloc[i]["tokens"]
    )
    jaccard_scores.append(score)

pred = pred.iloc[:min_length]
pred["jaccard_score"] = jaccard_scores

# ===============================
# Overall Score
# ===============================
average_jaccard = sum(jaccard_scores) / len(jaccard_scores)

print("📊 Jaccard Similarity Results")
print(f"Average Jaccard Similarity: {average_jaccard:.4f}")

pred.to_csv("jaccard_evaluation_output.csv", index=False, encoding="utf-8-sig")
print("💾 Saved as jaccard_evaluation_output.csv")
```
